```python
from langchain_community.chat_models import ChatOllama

llm = ChatOllama(model="llama3")

llm
```

```python
def multiply(a:int, b:int)->int:
    """this function will take two arguments a and b, just make their product"""
    result = a*b
    return result
    
```

```python
multiply(10,20)
```

```python
from langchain.tools import tool

@tool
def multiply(a:int, b:int)->int:
    """this function will take two arguments a and b, just make their product"""
    result = a*b
    return result
    
```

```python
multiply.invoke({'a':20, 'b':25})
```

```python
llm_with_tool = llm.bind_tools([multiply])
llm_with_tool
```
