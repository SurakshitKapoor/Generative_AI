```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="gemma:2b")

embeddings
```

```python
from langchain_core.documents import Document

documents = [
    Document(
        page_content="The Ganga is the longest river in India, originating from the Gangotri Glacier and flowing into the Bay of Bengal.",
        metadata={"id": 1, "river": "Ganga"}
    ),
    Document(
        page_content="The Yamuna is a major tributary of the Ganga and flows through Delhi, playing a crucial role in North India's water supply.",
        metadata={"id": 2, "river": "Yamuna"}
    ),
    Document(
        page_content="The Brahmaputra flows through Arunachal Pradesh and Assam, known for its vast basin and frequent floods.",
        metadata={"id": 3, "river": "Brahmaputra"}
    ),
    Document(
        page_content="The Godavari is the second-longest river in India and is often called the Dakshin Ganga of South India.",
        metadata={"id": 4, "river": "Godavari"}
    ),
    Document(
        page_content="The Narmada flows westward from the Amarkantak plateau and empties into the Arabian Sea.",
        metadata={"id": 5, "river": "Narmada"}
    ),
]

documents
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=documents,        
    embedding=embeddings         
)

vectorstore
```

```python
retriever = vectorstore.as_retriever(search_kwargs={'k':2})
retriever
```

```python
query = "which rivers belongs to Himalayan region?"
```

```python
result = retriever.invoke(query)
result
```
