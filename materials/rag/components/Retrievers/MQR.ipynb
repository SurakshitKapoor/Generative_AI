```python
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings
from langchain_ollama import ChatOllama
from langchain.retrievers.multi_query import MultiQueryRetriever
```

```python
documents = [
    Document(
        page_content="The Ganga is the longest river in India and is considered sacred.",
        metadata={"river": "Ganga"}
    ),
    Document(
        page_content="The Yamuna flows through Delhi and joins the Ganga at Prayagraj.",
        metadata={"river": "Yamuna"}
    ),
    Document(
        page_content="The Brahmaputra is known for floods in Assam.",
        metadata={"river": "Brahmaputra"}
    ),
    Document(
        page_content="The Godavari is the longest river in southern India.",
        metadata={"river": "Godavari"}
    ),
    Document(
        page_content="The Narmada is a west-flowing river flowing into the Arabian Sea.",
        metadata={"river": "Narmada"}
    ),
]
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")

embeddings
```

```python
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embeddings
)
```

```python
base_retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}
)

base_retriever
```

```python
llm = ChatOllama(model="gemma:2b")

llm
```

```python
mqr = MultiQueryRetriever.from_llm(
    retriever=base_retriever,
    llm=llm
)

mqr
```

```python
query = "Major rivers in northern India"

results = mqr.invoke(query)

for i, doc in enumerate(results, 1):
    print(f"\nResult {i}:")
    print(doc.page_content)
```
