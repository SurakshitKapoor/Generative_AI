```python
from langchain_ollama import OllamaEmbeddings

model = OllamaEmbeddings(model = "gemma:2b")

model
```

```python
from langchain_community.vectorstores import Chroma
```

```python
from langchain_core.documents import Document

doc1 = Document(
    page_content="The Ganga is the longest river in India, flowing through Uttarakhand, Uttar Pradesh, Bihar, and West Bengal.",
    metadata={"id": 1, "river": "Ganga"}
)

doc2 = Document(
    page_content="The Yamuna is a major tributary of the Ganga and flows through Delhi, Haryana, and Uttar Pradesh.",
    metadata={"id": 2, "river": "Yamuna"}
)

doc3 = Document(
    page_content="The Brahmaputra flows through Assam and is known for its vast basin and frequent flooding.",
    metadata={"id": 3, "river": "Brahmaputra"}
)

documents = [doc1, doc2, doc3]

documents
```

```python
#. Create vector store
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=model,
    collection_name="indian_rivers"
)
```

```python
vectorstore.get(include=['documents', 'embeddings', 'metadatas'])
```

```python
vectorstore.similarity_search(
    query="which river flows in assam?",
    k=1)
```

```python
vectorstore.similarity_search(
    query="which river flows in assam?",
    k=2)
```

```python
vectorstore.similarity_search_with_score(
    query="which river flows in assam?",
    k=1)
```
