```python
from langchain_ollama import OllamaEmbeddings
from langchain_experimental.text_splitter import SemanticChunker
```

```python
embeddings = OllamaEmbeddings(
    model="gemma:2b"
)
```

```python
semantic_splitter = SemanticChunker(
    embeddings=embeddings
)
```

```python
text = """
Cricket plays a central role in Indian culture and national identity.
From local street matches to international tournaments, the sport connects people across regions and generations.
Major victories by the national team often become shared moments of celebration.

The Indian Premier League changed how cricket is played, watched, and funded.
Shorter match formats, franchise ownership, and global broadcasting turned cricket into a commercial entertainment industry.
Players gained financial independence and global recognition through the league.

Modern cricket relies heavily on data and performance analytics.
Teams analyze batting patterns, bowling variations, pitch conditions, and weather data before every match.
These insights help coaches make tactical decisions and optimize team selection.

Artificial intelligence has expanded the use of analytics in sports.
Machine learning models can predict player fatigue, injury risks, and performance consistency.
AI systems also assist umpires and referees with decision-making technologies.

Beyond sports, artificial intelligence is transforming multiple industries.
In healthcare, AI supports medical imaging, disease detection, and treatment planning.
In finance, intelligent systems detect fraud, manage risk, and automate trading strategies.

Education is also being reshaped by artificial intelligence.
AI-powered tutoring systems adapt lessons based on a student’s learning pace and strengths.
This personalization improves engagement and long-term learning outcomes.

"""
```

```python
docs = semantic_splitter.create_documents([text])
```

```python
for i, doc in enumerate(docs, 1):
    print(f"\n--- Chunk {i} ---")
    print(doc.page_content)
```
