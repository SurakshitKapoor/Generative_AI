```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

model = ChatOllama(model="gemma:2b")

model
```

```python
result = model.invoke("hey, tell me about yourself!")
print(result)
```

```python
print(result.content)
```

```python

# taking the topic from user
topic = input("Enter the topic: ")

# building the templates
template_1 = PromptTemplate(
    template="Hey, tell me a detailed paragraph on the topic: {topic}",
    input_variables=["topic"]
)

template_2 = PromptTemplate(
    template="Now, make a 3-line short summary of this text: {text}",
    input_variables=["text"]
)
```

```python
# Step 1: Generate detailed paragraph
prompt_1 = template_1.invoke({"topic": topic})
response = model.invoke(prompt_1)
print(response.content)
print()

# Step 2: Generate summary
prompt_2 = template_2.invoke({"text": response.content})
summary = model.invoke(prompt_2)
print(summary.content)
```

```python
# now, with string output parser
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
# using chains to implement Output parsers
chain = template_1 | model | parser | template_2 | model | parser

result = chain.invoke({"topic":topic})

print(result)
```
