```python
from langchain_ollama import ChatOllama

model = ChatOllama(model="gemma:2b")
model
```

```python
from langchain_core.output_parsers import JsonOutputParser
parser = JsonOutputParser()
```

```python
# running with a static prompt
# format_instructions -> gets values on runtime

from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    template="""
    Write about Albert Einstein in detail.
    {format_instructions}
    """,
    input_variables=[],
    partial_variables={
        "format_instructions": parser.get_format_instructions()
    }
)

prompt = template.invoke({})
print(prompt)
```

```python
normal_response = model.invoke("")
print(normal_response)
```

```python
# getting th json format parser output from model
chain = template | model | parser

response = chain.invoke({})
print(response)
```

```python
# dynamic prompts -> taking input from user
topic = input("Enter the topic: ...")

template_2 = PromptTemplate(
    template="""
    Write about 5 important facts on {topic} in detail.
    Return strictly valid JSON. Do NOT include markdown, explanations, or comments. 
    All numbers must be pure numbers or strings. Example: "Population": "1.2 billion".

    {format_instructions}
    """,
    input_variables=['topic'],
    partial_variables={
        "format_instructions": parser.get_format_instructions()
    }
)

prompt_2 = template_2.invoke({"topic":topic})
print(prompt_2)
```

```python
chain_2 = template_2 | model | parser

# Pass input as a dict matching input_variables
response = chain_2.invoke({"topic": topic})

print(response)
```
