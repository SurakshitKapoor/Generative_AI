```python
from langchain_ollama import ChatOllama

model = ChatOllama(model="llama3")

model
```

```python
from langchain_core.tools import tool

@tool
def multiply(a:int, b:int)->int:
    """"given two numbers a and b, just return their product!"""
    return a * b

```

```python
result = multiply.invoke({'a':3,'b':5})
result
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

## Tool Binding:

binding the tool with llm, so that model gets introduced with the functuanilities of tool as well, and behave like that!

```python
llm_bind = model.bind_tools([multiply])
llm_bind
```

### Tool Calling:

process in which llm decides to call a tool while conversation / task.

LLM does not actually runs the tool. it just suggest to use that tool. then, tool can be used by langchain or you.

```python
llm_bind.invoke("how are you")
```
