***1. using @tool decorator***

```python
def multiply(a, b):
    return a * b
```

```python
def multiply(a: int, b: int) -> int:
    return a * b
```

```python
from langchain.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
result = multiply.run({"a": 4, "b": 5})
print(result)
```

```python
print(multiply.args)
print(multiply.name)
print(multiply.description)
```

***2. using Structured Pydantic Model:***

```python
from pydantic import BaseModel, Field

class MultiplyInput(BaseModel):
    a: int = Field(
        ...,
        description="First integer to multiply",
        ge=0
    )
    b: int = Field(
        ...,
        description="Second integer to multiply",
        ge=0
    )
```

```python
def multiply_numbers(a: int, b: int) -> int:
    return a * b
```

```python
from langchain_core.tools import StructuredTool

def multiply_numbers(a: int, b: int) -> int:
    return a * b

multiply_tool = StructuredTool.from_function(
    func=multiply_numbers,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput,
)

# Call
result = multiply_tool.run({"a": 3, "b": 4})
print(result)
```

```python
result = multiply_tool.run({"a": 6, "b": 7})
print(result)
```

***3. using Base Tool:*** 

```python
from pydantic import BaseModel, Field

class MultiplyInput(BaseModel):
    a: int = Field(..., description="First number", ge=0)
    b: int = Field(..., description="Second number", ge=0)
```

```python
from typing import Type
from langchain_core.tools import BaseTool

class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"
    args_schema: Type[BaseModel] = MultiplyInput  

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
tool = MultiplyTool()
result = tool.run({"a": 8, "b": 9})
print(result)
```

```python
print(tool.args)
print(tool.name)
print(tool.description)
```
