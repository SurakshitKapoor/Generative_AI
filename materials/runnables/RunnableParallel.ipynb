***Building a RunnableParallel***

```python
from langchain_ollama import ChatOllama

model = ChatOllama(model="gemma:2b")

model
```

```python
from langchain_core.prompts import PromptTemplate

prompt1 = PromptTemplate(
    template = """
        write a linkedin post on the topic of: {topic}
    """,
    input_variable=['topic']
)

prompt1
```

```python

prompt2 = PromptTemplate(
    template = """
        write a tweet on the topic of: {topic}
    """,
    input_variable=['topic']
)

prompt2
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

parser
```

using RunnableSequence or chains to create the sequences which runs parallely, 

```python
from langchain_core.runnables import RunnableParallel, RunnableSequence

chain = RunnableParallel( {
    'linkedin': RunnableSequence(prompt1, model, parser),
    'tweet': RunnableSequence(prompt2, model, parser) 
})

chain
```

```python
user_input = input("enter topic to search...\n")
```

```python
response = chain.invoke({"topic":user_input})

response
```

```python
linkedin = response['linkedin']
print("LinkedIN Post is: ", linkedin)

print()
print('*-'*120)
print()

tweet = response['tweet']
print("Tweeter post is: ", tweet)
```

```python
chain.get_graph().print_ascii()
```
